# Funciones de activación

**Objetivo**
Entender cómo una neurona transforma su salida al aplicar distintas funciones de activación

Una neurona recibe una entrada *x*, la multiplica por un peso *w*, le suma un sesgo
*b*, y obtiene una salida <br>
*y = w⋅x + b*. Después, pasa ese valor por una función de activación, que decide si “se activa” o no.

```python
# Importar librerías
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# Función para visualizar datos con y sin función de activación
def visualising_activation_function(x,z,y, act_function:str):
    plt.figure(figsize=(6,4))
    plt.title(f"Efecto de función de activación {act_function}")
    plt.plot(x, z, 'b--', label='Datos sin función de activación',linewidth=1)
    plt.plot(x, y, linewidth=2, label=f'Datos con función de activación {act_function}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Funciones de activación

#### Función de activación escalón  

$f(x) = \begin{cases} 0 & \text{si } x < 0 \\ 1 & \text{si } x \geq 0 \end{cases}$

```python
def step(x):
    out = np.where(x >= 0, 1, 0)
    return out
```

#### Función de activación sigmoide
# $f(x) = \frac{1}{1 + e^{-x}} $

```python
def sigmoid(x):
    out = 1 / (1 + np.exp(-x))
    return out
```

✍ **Actividad**  <br>                          Implementa las siguientes funciones de activación utilizando NumPy.
Cada función debe recibir un arreglo *x* y devolver otro arreglo con los valores transformados según las fórmulas dadas.

#### Función de activación tangente hiperbólica

# $f(x) = \frac{e^{x}-e^{-x}}{e^{x} + e^{-x}} $

```python
def tanh(x):
    """
    Calcula la tangente hiperbólica de x.
    """

    out = (np.exp(x) - np.exp(-x))/np.exp(x) + np.exp(-x)

    return out
```

#### Función de activación ReLU

$f(x) = \begin{cases} x & \text{si } x > 0 \\ 0 & \text{si } x \leq 0 \end{cases}$

```python
def relu(x):
    """
    Calcula la función ReLU.
    """
    out = np.maximum(0,x)

    return out
```

#### Función de activación Leaky ReLU

$f(x) = \begin{cases} x & \text{si } x > 0 \\ \alpha x  & \text{si } x \leq 0 \end{cases}$

Donde normalmente $\alpha=0.1$

```python
def leaky_relu(x, alpha=0.1):
    """
    Calcula la función Leaky ReLU.
    """
    out = np.where(x > 0, x, alpha * x)

    """
    Alternativo con Ifs
    if x > 0:
      out = x
    else:
      out = alpha * x
    """

    return out
```

# Verificar y visualizar las funciones de activación

```python
# Entradas (valores del -3 al 3)
x = np.linspace(-3, 3, 100)

# Parámetros de la neurona
w = 1.0
b = 0.0

# Salida lineal (sin activación)
z = w * x + b

print(z)
```

#### Función escalón

Manda a llamar las funciones que creaste y aplícalas sobre los datos de entrada x.
Después, utiliza la función visualising_activation_function() para visualizar los datos de entrada y cómo se transforman al aplicar una función de activación.   

~~~

~~~
Donde
- x → datos de entrada
- z → salida lineal (sin activación)

- 'nombre de la función' → el nombre de la función de activación usada (por ejemplo: 'sigmoide', 'tanh', 'ReLU')

```python
y_step = step(z)
visualising_activation_function(x,z,y_step,'Escalón')
```

#### Función sigmoide

```python
y_sigmoid = sigmoid(z)
visualising_activation_function(x,z,y_sigmoid,'Sigmoide')
```

#### Función tangente hiperbólica

```python
y_tanh = tanh(z)
visualising_activation_function(x,z,y_tanh,'Tangente Hiperbólica')
```

#### Función ReLU

```python
y_ReLU = relu(z)
visualising_activation_function(x,z,y_ReLU,'Rectified Linear Unit')
```

#### Función Leaky ReLU

```python
y_leaky_relu = leaky_relu(z)
visualising_activation_function(x,z,y_leaky_relu,'Rectified Linear Unit')
```
